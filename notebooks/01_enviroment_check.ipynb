DAC Test

```python

#imports
import librosa
import torch
import dac
import os
import soundfile as sf
import numpy

#CWD
os.chdir(r"F:\Text_to_Audio")
```

```python
#load audio
y, sr = librosa.load(r"test_audio\Banjo-0019.wav", sr = 44100, mono = True)

#audio to tensor
audio = torch.tensor(y).unsqueeze(0).unsqueeze(0) #shape: (1,1,samples)
#DAC expects 3D tensor
```

```python
#Load DAC model
model = dac.DAC.load(dac.utils.download())
model.eval().cuda()

#encode
with torch.no_grad():
    audio = audio.cuda()
    z, codes, latents, _, _ = model.encode(audio) #z = latent tensor, codes = discrete codebook indices, latents = pre-quantized continous features

#decode
with torch.no_grad():
    reconstructed = model.decode(z) #shape: (1,1,samples)

#export audio
audio_out = reconstructed.squeeze().cpu().numpy()
sf.write("dac_test.wav", audio_out, 44100)


```
